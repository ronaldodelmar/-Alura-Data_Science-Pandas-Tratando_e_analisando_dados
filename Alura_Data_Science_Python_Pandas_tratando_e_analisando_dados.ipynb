## Métodos de interpolação

```python
# importando a biblioteca
import pandas as pd
```

```python
# Criando um conjunto de dados para o exemplo
data = [0.5, None, None, 0,52, 0,54, None, None, 0.59, 0.6, None, 0.7]
s = pd.Series(data)
s
```

```python
# Podemos preencher com zeros ou outro valor 
s.fillna(0)
```

```python
# Mas as vezes queremos interpolar preenchendo sempre com valores que está acima (inclusive os preenchidos quando são mais de um )
s.fillna(method = 'ffill') 
```

```python
# Mas as vezes queremos interpolar preenchendo sempre com valores que está embaixo (inclusive os preenchidos quando são mais de um )
s.fillna(method = 'bfill') 
```

```python
# Preenchendo os valores faltantes com a média 
s.fillna(s.mean()) 
```

```python
# Podemos mesclar os valores limitando as ações, fazemos primeiro de cima pra baixo limitando em 1 substituição
s.fillna(method = 'ffill', limit = 1) 
```

```python
# Guardamos esse valor numa variável
s1 = s.fillna(method = 'ffill', limit = 1) 
```

```python
# Agora preenchemos com os dados faltantes de baixo para cima 
s1.fillna(method = 'bfill', limit = 1) 
```
